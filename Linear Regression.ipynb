```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.externals import joblib
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

## Only continuous features

```python
train_numerical = pd.read_csv('../data/processed/train_numerical.csv')
```

```python
LABEL = 'SalePrice'
FEATURES = list(train_numerical.columns)
FEATURES.remove(LABEL)
```

```python
training_set, validation_set = train_test_split(train_numerical, test_size=0.33, random_state=42)
training_set.head()
```

```python
lr_numerical = LinearRegression()
lr_numerical.fit(training_set[FEATURES], training_set[LABEL])
```

```python
predictions = lr_numerical.predict(validation_set[FEATURES])
print("Mean Squared Error", mean_squared_error(y_true=validation_set.SalePrice, y_pred=predictions))
```
